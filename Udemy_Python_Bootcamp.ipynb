```python
def new_recursion(n):
    if(n>0):
        result = n + new_recursion(n-1)
        print(result)
    else:
        result = 0
    return result
print("\n\n recursion results")
new_recursion(9)
```

## **SECTION 5 : Object-oriented programming model to organize software development**

Blueprint or prototype, bundling and functionality
- Class construction

```python
class NewClass:
    def newFunc(self):
        return 'python is fun'
        # objection creation
newObj = NewClass()
print(newObj.newFunc())
```

How to create multiple class objects

```python
class NewClass:
    def NewHello(self):
        return "hello from the NewHello function"
# first object
first_object = NewClass()
print(first_object.NewHello())
# second object
second_object = NewClass()
print(second_object.NewHello())
```

Multiple functions inside class

```python
class DoMath:
    def addition(self, x, y):
        return x + y
    def subtraction(self, x, y):
        return x - y
    def multiplication(self, x, y):
        return x * y
    def division(self, x, y):
        return x / y

# first object
first_object = DoMath()
print(first_object.addition(99, 96))
print(first_object.subtraction(99, 96))
print(first_object.multiplication(99, 96))
print(first_object.division(99, 96))

# second object
second_object = DoMath()
print(second_object.addition(4, 12))
print(second_object.subtraction(4, 12))
print(second_object.multiplication(4, 12))
print(second_object.division(4, 12))
```

Constructor in object oriented terminology

```python
class NewClass:
    def __init__(self):
        self.a = "hello"
    def Hi(self):
        return "hola, senorita"
first_object = NewClass()
print(first_object.a)
print(first_object.Hi)
```

Groups of tasks that belong to the object
